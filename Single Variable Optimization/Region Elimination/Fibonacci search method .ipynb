```python
#Fibonacci search - interval is given , no. of iterations are given , the interval is also given 
import math
from decimal import Decimal
def f(x):
    return Decimal((((x**3)-1)**3)+(((2*(x**2))-5)**4))
def F(n):
    if n<0:
        print("Incorrect input")
    elif n==0:
        return 1
    elif n==1:
        return 1
    else:
        return F(n-1)+F(n-2)
a=Decimal(input("enter the starting interval"))
b=Decimal(input("enter the ending interval"))
L=b-a
k=2
i=1
m=0
x1=0
x2=0
n=(input("what parameter are you going to use for evaluation : itn(Iteration Number) / (fnn)Function evaluation Number :"))
if n == "itn":
    u=Decimal(input("Choose a Iteration Number :"))
    m=u+1;
elif n == "fnn":
    u=Decimal(input("Choose a Function evaluation Number :"))
    m=u;
else:
    print ("Type a valid input as given in above question")
while True :
    print("")
    print("iteration no. = ", i)
    i=i+1
    print("k value is ", k)
    a1=(F((m-k)+1)/F(m+1))
    Lk=(Decimal(a1)*L)
    k=k+1
    x1=a+Lk
    x2=b-Lk
    fx1=f(x1)
    fx2=f(x2)
    print("")
    print("x1 =",x1,",x2 = ",x2)
    print("fx1 =",fx1,",fx2 =",fx2) 
    if fx1>fx2 :
        a=x1
    else:
        b=x2
    if k>m:
        print("Minimum Lies in the interval (",a,",",b,")")
        break
    else:
        continue
print("")
print("Total Number of iterations performed = ", i-1)
```
