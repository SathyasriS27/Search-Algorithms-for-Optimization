```python
#exhaustive search - interval , no. of intervals given 
import math
from decimal import Decimal
def f(x):
    return Decimal(x**2+(54/x))
a=Decimal(input("enter the starting interval"))
b=Decimal(input("enter the ending interval"))
n=Decimal(input("enter the number of intervals"))
dx=Decimal((b-a)/n)
x1=a
i=1
print("the length of the final interval = 2*dx =",2*dx)
print("")
while True :
    print("")
    print("iteration no. = ", i)
    i=i+1
    x2=x1+dx
    x3=x2+dx
    fx1=f(x1)
    fx2=f(x2)
    fx3=f(x3)
    check=fx1>=fx2 and fx2<=fx3
    print("")
    print("x1 =",x1,",x2 =",x2 ,",x3 = ",x3)
    print("fx1 =",fx1," ,fx2 =",fx2," ,fx3 =",fx3)
    print("Is fx1>fx2<fx3 :",check)
    print("")
    if check:
        print("Minimum Lies in the interval (",x1,",",x3,")")
        break
    else:
        print("fx1>fx2<fx3 is false")
        x1=x2
        x2=x3
        x3=x2+dx
        if x3>=b:
            print("no min")
            break
```
