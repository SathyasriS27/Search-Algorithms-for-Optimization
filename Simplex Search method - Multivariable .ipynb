```python
import math
from decimal import Decimal
def f(x,y):
    return ((x-1)**2+(y-2)**2)
x0=Decimal(input(("Enter the intital value of x0")))
y0=Decimal(input(("Enter the intital value of y0")))
x1=Decimal(input(("Enter the intital value of x1")))
y1=Decimal(input(("Enter the intital value of y1")))
x2=Decimal(input(("Enter the intital value of x2")))
y2=Decimal(input(("Enter the intital value of y2")))
gamma=Decimal(input(("Enter the intital value of gamma")))
beta=Decimal(input(("Enter the intital value of beta")))
J=Decimal(input("enter the number of iterations"))
i=1
while i<=J:
    fxx=[]
    print("")
    print("Iteration number:",i)
    fxx=[]
    fxx.append(f(x0,y0))
    fxx.append(f(x1,y1))
    fxx.append(f(x2,y2))
    print("x(0)=(",x0,",",y0,")\tf(x(0))=",fxx[0])
    print("x(1)=(",x1,",",y1,")\tf(x(1))=",fxx[1])
    print("x(2)=(",x2,",",y2,")\tf(x(2))=",fxx[2])
    fxx.sort()
    if fxx[0]==f(x0,y0):
        xb=x0
        yb=y0
    elif fxx[0]==f(x1,y1):
        xb=x1
        yb=y1
    elif fxx[0]==f(x2,y2):
        xb=x2
        yb=y2
    if fxx[1]==f(x0,y0):
        xg=x0
        yg=y0
    elif fxx[1]==f(x1,y1):
        xg=x1
        yg=y1
    elif fxx[1]==f(x2,y2):
        xg=x2
        yg=y2
    if fxx[2]==f(x0,y0):
        xw=x0
        yw=y0
    elif fxx[2]==f(x1,y1):
        xw=x1
        yw=y1
    elif fxx[2]==f(x2,y2):
        xw=x2
        yw=y2
    print("xb=(",xb,",",yb,")\tf(xb)=",fxx[0])
    print("xg=(",xg,",",yg,")\tf(xg)=",fxx[1])
    print("xw=(",xw,",",yw,")\tf(xw)=",fxx[2])
    xc=Decimal(0.5)*(xb+xg)
    yc=Decimal(0.5)*(yb+yg)
    f0=f(x0,y0)
    f1=f(x1,y1)
    f2=f(x2,y2)
    fc=f(xc,yc)
    print("xc=(",xc,",",yc,")\tf(xc)=",fc)
    xr=(2*xc)-xw
    yr=(2*yc)-yw
    fr=f(xr,yr)
    print("xr=(",xr,",",yr,")\tf(xc)=",fr)
    if fr<fxx[0]:
        print("f(xr)<f(xb)=",fr,"<",fxx[0])
        xnew=((1+gamma)*xc)-(gamma*xw)
        ynew=((1+gamma)*yc)-(gamma*yw)
    elif fr >= fxx[2]:
        print("f(xr)>=f(xw)=",fr,">=",fxx[2])
        xnew=((1-beta)*xc)+(beta*xw)
        ynew=((1-beta)*yc)+(beta*yw)
    elif fxx[1]<fr and fr<fxx[2]:
        print("f(xg)<f(xr)<f(xw)=",fxx[1],"<",fr,"<",fxx[2])
        xnew=((1+beta)*xc)-(beta*xw)
        ynew=((1+beta)*yc)-(beta*yw)
    fnew=f(xnew,ynew)
    print("xnew=(",xnew,",",ynew,")\tf(xnew)=",fnew)
    fxx.append(fnew)
    fxx.sort()
    if fxx[0]==f0:
        x00=x0
        y00=y0
    elif fxx[0]==f1:
        x00=x1
        y00=y1
    elif fxx[0]==f2:
        x00=x2
        y00=y2
    elif fxx[0]==fnew:
        x00=xnew
        y00=ynew
    if fxx[1]==f0:
        x11=x0
        y11=y0
    elif fxx[1]==f1:
        x11=x1
        y11=y1
    elif fxx[1]==f2:
        x11=x2
        y11=y2
    elif fxx[1]==fnew:
        x11=xnew
        y11=ynew
    if fxx[2]==f0:
        x22=x0
        y22=y0
    elif fxx[2]==f1:
        x22=x1
        y22=y1
    elif fxx[2]==f2:
        x22=x2
        y22=y2
    elif fxx[2]==fnew:
        x22=xnew
        y22=ynew
    i=i+1
    x0=x00
    y0=y00
    x1=x11
    y1=y11
    x2=x22
    y2=y22
    
    
```
