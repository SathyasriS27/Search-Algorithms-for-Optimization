```python
#newton raphson method - interval is given ,termination parameter is given 
import math
from decimal import Decimal
def f(x):
    return Decimal(((Decimal(math.sin(x))+(4*(x**2))))**2)
a=Decimal(input("enter the starting interval"))
E=Decimal(input("enter the termination parameter"))
i=1
while True :
    print("")
    print("iteration no. = ", i)
    print("a =",a)
    i=i+1
    if a < -0.01 or a > 0.01:
        dela=Decimal(0.01)*a
    elif a>=-0.01 and a<=0.01:
        dela=0.001
    else:
        print("invalid , try again")
    dela=round(dela,6)
    print("the value of del a is ",dela)
    diff_fa=(f(a+dela)-f(a-dela))/(2*dela)
    print("f'a =",diff_fa)
    diff2_fa=(f(a+dela)+f(a-dela)-(2*f(a)))/(dela**2)
    print("f\"a =",diff2_fa)
    a1=(a)-(diff_fa/diff2_fa)
    print("a1=",a1)
    if a1 < -0.01 or a1 > 0.01:
        dela1=Decimal(0.01)*a1
    elif a1>=-0.01 and a1<=0.01:
        dela1=0.001
    else:
        print("invalid , try again")
    dela=round(dela,6)
    diff_fa1=(f(a1+dela1)-f(a1-dela1))/(2*dela1)
    print("f'a1 =",diff_fa1)
    if abs(diff_fa1)<E:
        print("the minimum lies on ",a1)
        break
    else:
        a=a1
        continue

```
