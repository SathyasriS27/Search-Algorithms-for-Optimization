```python
#secant method - interval is given , termination parameter is given 
import math
from decimal import Decimal
def f(x):
    return Decimal(x**2+(54/x))
a=Decimal(input("enter the starting interval"))
b=Decimal(input("enter the ending interval"))
E=Decimal(input("enter the termination parameter"))
i=1
while True :
    print("")
    print("iteration no. = ", i)
    print("a =",a,",b=",b )
    if a < -0.01 or a > 0.01:
        dela=Decimal(0.01)*abs(a)
    elif a>=-0.01 and a<=0.01:
        dela=0.001
    else:
        print("invalid , try again")
    dela=round(dela,6)
    print("the value of del a is ",dela)
    if b < -0.01 or b > 0.01:
        delb=Decimal(0.01)*abs(b)
    elif b>=-0.01 and b<=0.01:
        delb=0.001
    else:
        print("invalid , try again")
    delb=round(delb,6)
    print("the value of del b is ",delb)
    diff_fa=(f(a+dela)-f(a-dela))/(2*dela)
    diff_fb=(f(b+delb)-f(b-delb))/(2*delb)
    print("f'a =",diff_fa," ,f'b =",diff_fb)
    if diff_fa<0 and diff_fb>0:
        print("the minimum lies between",a,"and",b)
    i=i+1
    z=b-(((b-a)/(diff_fb-diff_fa))*diff_fb)
    if z< -0.01 or z > 0.01:
        delz=Decimal(0.01)*z
    elif z>=-0.01 and z<=0.01:
        delz=0.001
    else:
        print("invalid , try again")
    delz=round(delz,6)
    print("the value of z=",z,",the value of delz =",delz) 
    diff_fz=(f(z+delz)-f(z-delz))/(2*delz)
    print("f'z =",diff_fz)
    if abs(diff_fz)<E:
        b=z
        print("the minimum lies on lies in the interval (",a,",",b,")")
        break
    elif diff_fz<0:
        a=z
        continue
    elif diff_fz>0:
        b=z
        continue
```
