```python
#HOOKE JEEVES FOR 2 VARIABLES , WITH TERMINATION PARAMETER 
import math
from decimal import Decimal
def f(x,y):
    return Decimal((x-1)**2+(y-2)**2)
x0=input(("Enter the intital value of x0"))
y0=input(("Enter the intital value of y0"))
delx=Decimal(input(("Enter the x value in the reduction parameter")))
dely=Decimal(input(("Enter the y value in the reduction parameter")))
E=Decimal(input(("enter the termination parameter")))
alpha=Decimal(input(("enter the reduction parameter or alpha")))
k=0
i=1
j=0
l=0
x1=[]
y1=[]
fminb=[]
fcomp=[]
while True :
    print("")
    print("iteration no. :",i)
    print("For X")
    x=[]
    y=[]
    fxx=[]
    fxy=[]
    x.append(Decimal(x0))
    y.append(Decimal(y0))
    f0=f(Decimal(x0),Decimal(y0))
    fxx.append(f0)
    x.append(Decimal(x[k]+(delx)))
    x.append(Decimal(x[k]-(delx)))
    fxx.append(f(x[1],y[0]))
    fxx.append(f(x[2],y[0]))
    p=0
    print("f=(",x0,",",y0,")\tf(x(0))=",f0)
    print("f+=(",x[1],",",y[0],")\tf(x(1))=",fxx[1])
    print("f-=(",x[2],",",y[0],")\tf(x(2))=",fxx[2])
    fmin=min(fxx)
    print("fmin=",fmin)
    fminb.append(fmin)
    if fxx[0]==fmin:
        x01=x0
        y01=y0
    elif fxx[1]==fmin:
        x01=x[1]
        y01=y0
    elif fxx[2]==fmin:
        x01=x[2]
        y01=y0
    print("the new values for calculation of Y is (",x01,",",y01,")")
    print("For Y")
    y.append(Decimal(y[k]+(dely)))
    y.append(Decimal(y[k]-(dely)))
    fxy.append(f(Decimal(x01),Decimal(y01)))
    fxy.append(f(Decimal(x01),y[1]))
    fxy.append(f(Decimal(x01),y[2]))
    print("f=(",x01,",",y01,")\tf(x(0))=",fxy[0])
    print("f+=(",x01,",",y[1],")\tf(x(1))=",fxy[1])
    print("f-=(",x01,",",y[2],")\tf(x(2))=",fxy[2])
    fmin=min(fxy)
    print("fmin=",fmin)
    fminb.append(fmin)
    if fxy[0]==fmin:
        x02=x01
        y02=y01
    elif fxy[1]==fmin:
        x02=x01
        y02=y[1]
    elif fxy[2]==fmin:
        x02=x01
        y02=y[2]
    x1.append(Decimal(x0))
    x1.append(Decimal(x02))
    y1.append(Decimal(y0))
    y1.append(Decimal(y02))
    print("min lies in (",x02,",",y02,")")
    fcomp.append(f(x02,y02))
    if l>=1:
        if (fcomp[i-1]<fcomp[i-2]):
            print("f(x(",i,"))<f(x(",i-1,"))")
            
        elif (fcomp[i-1]==fcomp[i-2]):
            print("f(x(",i,"))=f(x(",i-1,"))")
            moddel=((delx)**2+(dely)**2)**Decimal(0.5)
            if moddel>E:
                p=1
                print("||del||>E=||",moddel,"||>",E)
                delx=delx/2
                dely=dely/2
                print("the new values of delx and dely are =(",delx,",",dely,")")
                x0=x02
                y0=y02
                i=i+1
                j=j+2
                l=l+1
                continue
        else :
            print("f(x(",i,"))>f(x(",i-1,"))") 
    if (x0==x02) and (y0==y02):
        print("Exploratory move is not a success")
        moddel=((delx)**2+(dely)**2)**Decimal(0.5)
        if moddel<E:
            print("||del||<E=||",moddel,"||<",E)  
            print("min lies in (",x02,",",y02,")")
            break
    elif p!=1:
        print("Exploratory move is a success")
        print("pattern move :")
        if j==0:
            xp1=2*(x1[j+1])-(x1[j])
            yp1=2*(x1[j+1])-(x1[j])
            print("x(p)=(",xp1,",",yp1,")")
            
        else :
            xp1=2*(x1[j+1])-(x1[j-1])
            yp1=2*(y1[j+1])-(y1[j-1])
            print("x(p)=(",xp1,",",yp1,")")
        x0=xp1
        y0=yp1
        i=i+1
        j=j+2
        l=l+1
        
        
```
